# Elbow Method

```python
# Imports

import os
import sys

sys.path.append(os.path.abspath(".."))

from src.preprocessing import process_X
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# Data Load

df = pd.read_csv(r"C:\Users\sailj\OneDrive\文档\GitHub\Data-Driven Customer Segmentation\dataset\Online Retail.csv")

X = process_X(df)

X
```

```python
# Training Models with different k

wcss = []
score = []


for k in range(2,11) :
    model = KMeans(
        n_clusters = k ,
        n_init = 10 ,
        random_state = 42
    )
    
    model.fit(X)
    wcss.append(model.inertia_)
    label = model.labels_
    score.append(silhouette_score(X , label))
```

```python
# Elbow Curve 

plt.plot(range(2,11) , wcss)
plt.xlabel("no of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Curve")
plt.grid()
plt.show()
```

```python
# Silhouette Score Curve 

plt.plot(range(2,11) , score)
plt.xlabel("no of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score Curve")
plt.grid()
plt.show()
```

# Final Verdict
